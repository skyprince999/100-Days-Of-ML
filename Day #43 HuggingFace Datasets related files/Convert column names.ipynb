```python
%%time
for filename in os.listdir():
    if '.json' in filename:
        print("Processing file: ", filename)
        dutch_tweets = pd.read_json(filename)
        print(dutch_tweets.shape)
        print(dutch_tweets.columns)
        colnames = list(dutch_tweets.columns)
        
        #Drop the Day column
        if "Day" in colnames:
            dutch_tweets.drop(['Day'], inplace=True, axis=1)
            colnames.remove('Day')
            
        colnames = [col.lower().replace(" ", "_").replace(".nl", "") for col in colnames]
        dutch_tweets.columns = colnames
        
        dutch_tweets = dutch_tweets[['full_text', 'text_translation', 'created_at', 'screen_name', 'description', 'desc_translation',
                                     'weekofyear', 'weekday', 'day', 'month', 'year',
                                     'location', 'point_info', 'point', 'latitude', 'longitude', 'altitude', 'province', 
                                     'hisco_standard', 'hisco_code', 'industry',
                                     'sentiment_pattern', 'subjective_pattern']]
        
        #convert to json format
        result = dutch_tweets.to_json(orient="records")
        parsed = json.loads(result)

        # Create the json filename
        file_comp = list()
        file_comp.append("dutch-tweets")
        file_comp.append(filename)
        json_filename = "/".join(file_comp)
        
        print("Output file: ", json_filename)
        with open(json_filename, 'w') as outfile:
            json.dump(parsed, outfile)
        
        #Convert created_at and object cols to string 
        
        #Write the excel file
#         excel_filename = filename[:-4] + 'xlsx'
#         dutch_tweets.to_excel(excel_filename, index=False)
        
        
        print("####################################################")
        print("####################################################")
        
```
