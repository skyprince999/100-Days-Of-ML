```python
import os 
import pandas as pd
import numpy as np

filenames = list()
for file in os.listdir("/home/ubuntu"):
    if file.endswith("json"):
        filenames.append(file)
        
print(filenames)
```

```python
dutch_tweets = pd.read_json(filenames[0])
print(dutch_tweets.shape)
dutch_tweets.head()
```

```python
for file in filenames:
    print(f"Reading file...{file}")
    dutch_tweets = pd.read_json(file)
    print("Writing to excel...")
    print(f"Shape of you...{dutch_tweets.shape}")
    
    excel_file = file[:-4] + "xlsx"
    dutch_tweets.to_excel(excel_file, index=False)
```
