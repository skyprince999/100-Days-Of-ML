```python
# WRITE (COPY) SAMPLE FILES TO ALL FOLDERS
for pair in language_pair:
    source = pair[0]
    target = pair[1]
    
    config_name = source+ "_" + target + "/"
    
    config_dir = base_dir+config_name
    
    if not os.path.exists(config_dir):
        os.makedirs(config_dir)
        os.makedirs(config_dir+version)
        os.makedirs(config_dir+version + "dummy_data/")
        os.makedirs(config_dir+version + "dummy_data/"+train)
        os.makedirs(config_dir+version + "dummy_data/"+val)
        os.makedirs(config_dir+version + "dummy_data/"+test)
        
    source_file_data = "sample_en.zip"
    target_file_data = "sample_hi.zip"
    
    source_file_dummy = config_dir+version + "dummy_data/"+train + "ted_" + source + "-20140120.zip"
    target_file_dummy = config_dir+version + "dummy_data/"+train + "ted_" + target + "-20140120.zip"
    
    if not os.path.exists(source_file_dummy) and not os.path.exists(target_file_dummy):
        copyfile(source_file_data, source_file_dummy)
        copyfile(target_file_data, target_file_dummy)
    
    source_file_dummy = config_dir+version + "dummy_data/"+val + "ted_" + source + "-20150530.zip"
    target_file_dummy = config_dir+version + "dummy_data/"+val + "ted_" + target + "-20150530.zip"
    
    if not os.path.exists(source_file_dummy) and not os.path.exists(target_file_dummy):
        copyfile(source_file_data, source_file_dummy)
        copyfile(target_file_data, target_file_dummy)
        
    source_file_dummy = config_dir+version + "dummy_data/"+test + "ted_" + source + "-20160408.zip"
    target_file_dummy = config_dir+version + "dummy_data/"+test + "ted_" + target + "-20160408.zip"
    
    if not os.path.exists(source_file_dummy) and not os.path.exists(target_file_dummy):
        copyfile(source_file_data, source_file_dummy)
        copyfile(target_file_data, target_file_dummy)
    
```

```python
#ZIP FILES IN DUMMY_DATA folder
for pair in language_pair:
    
    source = pair[0]
    target = pair[1]
    config_name = source+ "_" + target + "/"
    config_dir = base_dir+config_name+version
    
    shutil.make_archive(config_dir+"dummy_data", 'zip', config_dir+"dummy_data")
    
```

```python
#DELETE ALL DUMMY_DATA folders 
for pair in language_pair:
    
    source = pair[0]
    target = pair[1]
    config_name = source+ "_" + target + "/"
    config_dir = base_dir+config_name+version
    
    shutil.rmtree(config_dir + "dummy_data/")
```
