```python
tweet = dutch['full_text'][1]
sentiment(tweet) #'Een onwijs spannend goed boek!'
```

```python
senti_pattern = list()
subj_pattern = list()

for idx, row in dutch.iterrows():
    try:
        senti = sentiment(row['full_text'])
        senti_pattern.append(senti[0])
        subj_pattern.append(senti[1])
    except:
        senti_pattern.append(-1)
        subj_pattern.append(-1)
        
dutch["Sentiment_pattern.nl"] = senti_pattern
dutch["Subjective_pattern.nl"] = subj_pattern
```
