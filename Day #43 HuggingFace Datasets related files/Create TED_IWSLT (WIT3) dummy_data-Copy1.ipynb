```python
from __future__ import absolute_import, division, print_function

import os
import xml.etree.ElementTree as ET
import zipfile
from collections import defaultdict

import datasets
import shutil
from shutil import copyfile

# shutil.make_archive(output_filename, 'zip', dir_name)
```

```python
base_dir = "wit/dummy/"
data_dir = "wit/XML_releases/"

version = "1.1.0/"
```

```python
# Year subscripts for the specific folder
_YEAR = {
    "2014": "-20140120",
    "2015": "-20150530",
    "2016": "-20160408"
}

_YEAR_FOLDER = {
    "2014": "XML_releases/xml-20140120/",
    "2015": "XML_releases/xml-20150616/",
    "2016": "XML_releases/xml/"
}

source_file_data = "sample_en.xml"
target_file_data = "sample_hi.xml"
```

### Copying the same XML files for each language pair

```python
# WRITE (COPY) SAMPLE FILES TO ALL FOLDERS
for pair in _LANGUAGE_PAIRS:
    for year in _YEAR.keys():

        source = pair[0]
        target = pair[1]

        config_name = source+ "_" + target +"_" + year+ "/"

        config_dir = base_dir+config_name

        if not os.path.exists(config_dir):
            os.makedirs(config_dir)
            os.makedirs(config_dir+version)
            os.makedirs(config_dir+version + "dummy_data/")
            os.makedirs(config_dir+version + "dummy_data/dummy_data/")
            os.makedirs(config_dir+version + "dummy_data/dummy_data/"+_YEAR_FOLDER[year])
            
        
        source_file_dummy_xml = "ted_" + source + _YEAR[year] + ".xml"
        target_file_dummy_xml = "ted_" + target + _YEAR[year] + ".xml"
        
        source_file_dummy_zip = "ted_" + source + _YEAR[year] + ".zip"
        target_file_dummy_zip = "ted_" + target + _YEAR[year] + ".zip"
        
        
        copyfile(source_file_data, source_file_dummy_xml)
        copyfile(target_file_data, target_file_dummy_xml)
            
        zipfile.ZipFile(source_file_dummy_zip, mode='w').write(source_file_dummy_xml)
        zipfile.ZipFile(target_file_dummy_zip, mode='w').write(target_file_dummy_xml)
        
        source_file_dummy = config_dir+version + "dummy_data/dummy_data/"+_YEAR_FOLDER[year] + "ted_" + source + _YEAR[year] + ".zip"
        target_file_dummy = config_dir+version + "dummy_data/dummy_data/"+_YEAR_FOLDER[year] + "ted_" + target + _YEAR[year] + ".zip"
        
        if not os.path.exists(source_file_dummy) and not os.path.exists(target_file_dummy):
            copyfile(source_file_dummy_zip, source_file_dummy)
            copyfile(target_file_dummy_zip, target_file_dummy)
        
        os.remove(source_file_dummy_zip)
        os.remove(target_file_dummy_zip)
        
        os.remove(source_file_dummy_xml)
        os.remove(target_file_dummy_xml)
```

```python
#ZIP FILES IN DUMMY_DATA folder
for pair in _LANGUAGE_PAIRS:
    for year in _YEAR.keys():
        source = pair[0]
        target = pair[1]

        config_name = source+ "_" + target +"_" + year+ "/"

        config_dir = base_dir+config_name+version

        shutil.make_archive(config_dir+"dummy_data", 'zip', config_dir+"dummy_data")
```

```python
#DELETE ALL DUMMY_DATA folders 
for pair in _LANGUAGE_PAIRS:
    for year in _YEAR.keys():
        source = pair[0]
        target = pair[1]

        config_name = source+ "_" + target +"_" + year+ "/"

        config_dir = base_dir+config_name+version

        shutil.rmtree(config_dir + "dummy_data/")
```
